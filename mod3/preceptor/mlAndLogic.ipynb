```python
# initialising the value
n = 0.1
w0 = 0.5
w1 = 0.3
w2 = 0.4
i = 0
x0 = 1
op = 0.0
x1 = 0
x2 = 0
def updateWeight(row, o):
    global w0, w1, w2, x0
    t = 0.5
    dw0 = 0
    dw1 = 0
    dw2 = 0
    
    dw0 = n *(t - o) * x0
    dw1 = n *(t - o) * x1
    dw2 = n *(t - o) * x2
    
    w0 = w0 + dw0
    w1 = w1 + dw1
    w2 = w2 + dw2
    
    print(w0, w1, w2)
    iteration(row)
    
def iteration(rowVal):
    print('row value ',rowVal)
    print('\n')
    y1 = 0
    x1 = int(rowVal['x1'])
    x2 = int(rowVal['x2'])
    y = int(rowVal['y'])
    print("1] initial value\nx0 x1 x2 y y1")
    print(x0, x1, x2, y, y1)
        
#         step function
    op = float((x0 * w0) + (x1 * w1) + (x2 * w2))
    print('2] step function op value ', op)
#     print('\n') 
    if op< 0.5:
        y1 = 0
    else:
        y1 = 1
            
    print("3] After step function value\nx0 x1 x2 y y1")
    print(x0, x1, x2, y, y1)
    if y1 != y:
        print('  i] not equal')
        updateWeight(rowVal, op)
            
    else:
        
        print('  ii] Going to check for next row')
        return 0
        
    
    
    
        
from csv import DictReader
with open('andLogicSheet.csv', 'r') as read_obj:
    count =0 
    data = DictReader(read_obj)
    for rowVal in data:
        count = count + 1
        print('*' * 50, '-', count ,'-', '*' * 50)
        iteration(rowVal)
        
 


print('\n',' '*5, '--'* 25)
print('\n')
print('final value')
print('w0 = ', w0)
print('w1 = ', w1)
print('w2 = ', w2)
#     print(type(data))
        
    
```
